```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def lucas_num(n):
    if n == 0:
        return 2
    if n == 1:
        return 1
    
    return lucas_num(n-2) + lucas_num(n-1)
```

```python
data = []
```

```python
for i in range(15):
    data.append(lucas_num(i))
```

```python
plt.xkcd()
```

```python
plt.plot(data)
```
